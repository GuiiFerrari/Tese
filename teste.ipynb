```python
import numpy as np
from lmfit.models import ExpressionModel
from lmfit import Parameters
import matplotlib.pyplot as plt

```

```python
x = np.linspace(0, 10, 20)
a = 5
b = 2
sy = 0.6
y = a*x + b + sy*np.random.randn(len(x))


ind_var = "neymar"
funcao_ajuste = f"a*{ind_var} + b"
f = ExpressionModel(funcao_ajuste, independent_vars=[ind_var])
params = Parameters()
params.add('a', value=1, vary=True)
params.add('b', value=1, vary=True)
result = eval(f"f.fit(data = y, params = params, {ind_var} = x)", None, {"f": f, "y": y, "params" : params})
print(result.fit_report())
fig = plt.figure(dpi = 180)
plt.scatter(x, y, s = 3, label='data')
plt.plot(x, result.best_fit, 'r-', label='best fit')
plt.legend(loc='best')
plt.show()
```
